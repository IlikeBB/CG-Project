```python
import matplotlib.pyplot as plt
import numpy as np, pandas as pd
```

```python
value_type = 'var' #var, mean
data_type = 'fc' #fc, alff
outcome_ = 'AD8主觀認知障礙[0-1,>=2]' #AD8主觀認知障礙[0-1,>=2], MOCA客觀認知測驗分數[<=23,>=24]
df = pd.read_csv('./feature_selection_result/{}_{}_{}.csv'.format(outcome_, value_type, data_type))
print(df.columns)
```

```python
basic = df.sort_values('rank')
basic['rank_value'] = (basic['rank']-max(basic['rank']))*-1+1
```

```python

filter_rank = 100
filter_basic = basic[basic['rank']<filter_rank]
basic = df.sort_values('rank')
filter_basic
```

```python
y = filter_basic['feature_location']

x = filter_basic['rank_value']
fig = plt.figure(figsize=(20,50))
plt.yticks(np.arange(len(y)), y[::-1])
plt.barh(y, (x-31000)[::-1])
plt.title("FC Feature Selection Top - {}".format(filter_rank))
plt.show()
```
