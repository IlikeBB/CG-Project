```python
import pandas as pd
import numpy as np
import math

compare_csv = pd.read_csv('../../csv/sorting_MRA_202101-202111+seq+O.csv')
# compare_csv = compare_csv[compare_csv['ManufacturerModelName']=='Skyra']
compare_csv = compare_csv[['PatientID', 'AccessionNumber']]
# compare_csv
```

```python
raw_csv = pd.read_csv('../../csv/350~nihssmrs.csv')
raw_csv = raw_csv[raw_csv['3T MRI']!=0]
raw_csv = raw_csv[raw_csv['3T MRI']!=1.0]
# raw_csv
```

```python
new_name_csv = []
for o, (i,j) in enumerate(zip(raw_csv['ID'], raw_csv['病歷號碼'])):
    # print([o,i,j])
    for x,y in zip(compare_csv['PatientID'], compare_csv['AccessionNumber']):
            if int(j)==int(x):
                # print(int(j), int(x))
                new_name_csv.append([i, '-', int(j), int(x), y])
pd_new_name = pd.DataFrame(new_name_csv, columns=['ID', '1.5T&3.0T MRI','raw_patient_id', 'compare_patient_id', 'AccessionNumber'])
pd_new_name
```
